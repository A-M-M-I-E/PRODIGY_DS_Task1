```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
%matplotlib inline

data_url = 'C:/Users/ameya/Downloads/countries-table.csv'
df = pd.read_csv(data_url)

# Data preprocessing
# Remove rows with missing values
df = df.dropna()  
df.head()
```

```python
top_10_countries = df.nlargest(10, 'pop2023')
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='pop2023', data=top_10_countries)
plt.title('10 Countries by Population by (2023)')
plt.xlabel('Countries')
plt.ylabel('Population ')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.histplot(data=df, x='growthRate', bins=20, kde=True)
plt.title('Population Growth  Distribution')
plt.xlabel('Growth Rate ')
plt.ylabel('Number ')
plt.show()
```
